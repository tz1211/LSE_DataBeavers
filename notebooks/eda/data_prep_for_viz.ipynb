```python
import pandas as pd 
import geopandas as gpd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("../../data/bike_car_ped_accidents_2021.csv")

gdf = gpd.read_file("../../data/bike_lanes_Berlin.geojson")

gdf.plot()
plt.show()
```

# Union join accidents and bike lane for Tableau plotting 

```python
df_merged = pd.concat([df, gdf], ignore_index=True)
df_merged
```
